```python
data2 = data.copy()   #make a copy of the dataframe

#Replace all the outliers with median values. This will create some new outliers,, but we will ignore them for the time being

for i, j in zip(np.where(z > threshold)[0], np.where(z > threshold)[1]):# iterate using 2 variables.i for rows and j for columns
    data.iloc[i,j] = data.iloc[:,j].median()  # replace i,jth element with the median of j i.e, corresponding column
z = np.abs(stats.zscore(data2))
data.shape
```

```python
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
std_scale
```

```python
data['Parameter 1_scaled'] = std_scale.fit_transform(data[['Parameter 1']])
data['Parameter 2_scaled'] = std_scale.fit_transform(data[['Parameter 2']])
data['Parameter 3_scaled'] = std_scale.fit_transform(data[['Parameter 3']])
data['Parameter 4_scaled'] = std_scale.fit_transform(data[['Parameter 4']])
data['Parameter 5_scaled'] = std_scale.fit_transform(data[['Parameter 5']])
data['Parameter 6_scaled'] = std_scale.fit_transform(data[['Parameter 6']])
data['Parameter 7_scaled'] = std_scale.fit_transform(data[['Parameter 7']])
data['Parameter 8_scaled'] = std_scale.fit_transform(data[['Parameter 8']])
data['Parameter 9_scaled'] = std_scale.fit_transform(data[['Parameter 9']])
data['Parameter 10_scaled'] = std_scale.fit_transform(data[['Parameter 10']])
data['Parameter 11_scaled'] = std_scale.fit_transform(data[['Parameter 11']])
```

```python
y=data['Signal_Strength']
x=data[['Parameter 1', 'Parameter 2', 'Parameter 3', 'Parameter 4', 'Parameter 5', 'Parameter 6', 'Parameter 7', 'Parameter 8', 'Parameter 9', 'Parameter 10', 'Parameter 11']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=1)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
import tensorflow as tf
tf.__version__
```

```python
# Initialize the random number generator
import random
random.seed(0)

# Ignore the warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
model = tf.keras.models.Sequential() #Initialise the sequential model
```

```python
#Initializers
#initializer = tf.keras.initializers.RandomNormal(mean=0., stddev=1.)
initializer = tf.keras.initializers.GlorotNormal()
model.add(tf.keras.layers.Dense(50, kernel_initializer=initializer))     # use he_normal initializer
#model.add(tf.keras.layers.Dense(50, kernel_initializer=initializer))     # use Xavier_normal initializer
```

```python
model.add(tf.keras.layers.Dense(11, input_dim = x_train.shape[1], kernel_initializer=initializer))
```

```python
#Normalize the data
model.add(tf.keras.layers.BatchNormalization())
```

```python
model.add(tf.keras.layers.Dense(11, kernel_initializer=initializer,activation='relu'))
```

```python
#model.add(tf.keras.layers.Dense(1))
model.add(tf.keras.layers.Dense(1, kernel_initializer=initializer,activation='linear', kernel_regularizer=tf.keras.regularizers.l2(0)))
```

```python
sgd = tf.keras.optimizers.SGD(learning_rate=0.001, decay=1e-6, momentum=0.9)
sgd_optimizer = tf.keras.optimizers.SGD(learning_rate=0.001, momentum=0.0, nesterov=False, name='SGD')
adam_optimizer = tf.keras.optimizers.Adam(learning_rate=0.00001, beta_1=0.9, beta_2=0.999, epsilon=1e-07, amsgrad=True,name='Adam')
RMSprop_optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.00001, rho=0.8, momentum=0.09, epsilon=1e-07, centered=False, name='RMSprop')
Adagrad_optimizer = tf.keras.optimizers.Adagrad(learning_rate=0.001, initial_accumulator_value=0.1, epsilon=1e-07,name='Adagrad')
model.compile(optimizer = sgd, loss='mean_squared_error', metrics = ['mean_absolute_error'])
```

```python
# Use earlystopping
callback = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=2, min_delta=0.01)
model.fit(x_train,y_train,epochs=100,validation_data=(x_test,y_test), batch_size=100)
#model.fit(x_train,y_train,epochs=100,validation_data=(x_test,y_test), batch_size=100, callbacks=[callback])
```
